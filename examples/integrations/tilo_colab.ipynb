## 2. See a surface render inline (no API key needed)

`tilo.notebook(spec)` renders any AIP spec inline.

```python
import tilo

spec = {
    "version": "tilo/aip/v1",
    "title": "Contract Risk Review",
    "status": "ready",
    "blocks": [
        {"id": "h", "type": "heading", "props": {"text": "2 High-Risk Clauses Found", "severity": "high"}},
        {"id": "m1", "type": "metric", "props": {"label": "Risk Score", "value": "7.2", "delta": "+1.1"}},
        {"id": "m2", "type": "metric", "props": {"label": "Clauses", "value": "24"}},
        {"id": "chart", "type": "chart", "title": "Risk by Category",
         "props": {"chart_type": "radar", "axes": [
             {"label": "Liability", "score": 9}, {"label": "Payment", "score": 6},
             {"label": "IP", "score": 8}, {"label": "Termination", "score": 3}]}},
        {"id": "cl", "type": "checklist", "props": {"items": [
            {"text": "Review liability cap", "checked": True},
            {"text": "Confirm payment terms"}]}},
        {"id": "conf", "type": "confirmation", "props": {"description": "Approve revised contract?", "risk_level": "high"}},
        {"id": "mem", "type": "memory_card", "props": {"content": "User prefers conservative liability caps", "confidence": 0.85}},
    ],
    "views": [
        {"id": "v1", "label": "Risks", "block_ids": ["h", "m1", "m2", "chart"]},
        {"id": "v2", "label": "Decision", "block_ids": ["cl", "conf", "mem"]},
    ],
    "follow_ups": ["Compare to industry standard", "Draft a counter-proposal"],
}

tilo.notebook(spec)
```

## 3. Generate one from your own LLM

Set your `OPENAI_API_KEY`, then let the model author the surface.

```python
import os
os.environ["OPENAI_API_KEY"] = "sk-..."  # or set it in Colab secrets

import tilo

spec = tilo.generate(
    "Analyse this quarter's sales pipeline and recommend the top 3 follow-up actions.",
    model="gpt-4o-mini",
    skill="sales_dashboard",
)

print(spec.title)
print([b.type for b in spec.blocks])
tilo.notebook(spec)
```
